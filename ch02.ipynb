```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

#2.3 多変量正規分布
$\mathcal{N}(\mathbb{x}\mid\mathbb{\mu},\mathbb{\Sigma})=\dfrac{1}{(2\pi)^D\sqrt{|\mathbb{\Sigma}|}}\mathrm{exp}\Big(-\dfrac{1}{2}(\mathbb{x}-\mathbb{\mu})^T\mathbb{\Sigma}^{-1}(\mathbb{x}-\mathbb{\mu})\Big)$

これをNumPyで実装してみる

```python
mu=np.array([0,0]) #平均ベクトル
sigma1=np.array([[1,0.9],[0.9,1]]) #共分散行列
sigma2=np.array([[1,-0.7],[-0.7,1]])
sigma3=np.array([[1,0],[0,1]])
n=500
x1=np.random.multivariate_normal(mu,sigma1,n) #乱数をn個生成 shape:(n,2)
x2=np.random.multivariate_normal(mu,sigma2,n)
x3=np.random.multivariate_normal(mu,sigma3,n)


fig,axes=plt.subplots(1,3,figsize=(20,5))
axes[0].scatter(x1[:,0],x1[:,1])
axes[1].scatter(x2[:,0],x2[:,1])
axes[2].scatter(x3[:,0],x3[:,1])

for ax in axes:
  ax.spines['left'].set_position('zero')    
  ax.spines['right'].set_position('zero')
  ax.spines['bottom'].set_position('zero')
  ax.spines['top'].set_position('zero')
  ax.grid(True)
  ax.set_xlabel("x",loc="right") 
  ax.set_ylabel("y",loc="top",rotation="horizontal")



plt.show()
```
